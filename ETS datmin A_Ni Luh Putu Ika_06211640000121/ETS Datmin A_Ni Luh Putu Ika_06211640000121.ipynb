```python
online['Description'] = online['Description'].str.strip()
online.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
  
online['InvoiceNo'] = online['InvoiceNo'].astype('str') 
online = online[~online['InvoiceNo'].str.contains('C')]

online['UnitPrice'] = online['UnitPrice'].astype('str')
online = online[~online['UnitPrice'].str.contains('0')]
```

```python
online.shape
```

```python
basket = online.groupby('InvoiceNo')
invoice = []
for i,j in basket:
    invoice.append(list(j['Description'].map(str)))
```

```python
data = pd.DataFrame(invoice)
```

```python
data.shape
```

```python
records=[]
for i in range (0,19168):
    records.append([str(data.values[i,j]) for j in range(0,923) if str(data.loc[i,j])!="None"])
```

```python
records
```

```python
association_rules = apriori(records, min_support=0.0045,
                           min_confidence=0.2,
                           min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result:
    pair=item[0]
    items = [x for x in pair]
    print("rule: "+items[0]+"->"+items[1])
    print("support: "+str(item[1]))
    print("confidence: "+str(item[2][0][2]))
    print("lift: "+str(item[2][0][3]))
    print("=========================================")
```

```python
basket2 = (online.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket2.head()
```

```python
def encode_units(x1):
    if x1 <= 0:
        return 0
    if x1 >= 1:
        return 1
basket_sets = basket2.applymap(encode_units)
 
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets.head()
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
frequent_itemsets.head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
