```python
import pandas as pd
import numpy as np
```

```python
adult = pd.read_csv('adult_test.csv',sep=';')
adult.head(10)
```

```python
adult.describe()
```

```python
np.sum(adult.isnull())
```

```python
from scipy.stats import mode

adult['Workclass'] = adult['Workclass'].fillna(adult['Workclass']).dropna().mode()[0]
adult['Occupation'] = adult['Occupation'].fillna(adult['Occupation']).dropna().mode()[0]
adult['Country'] = adult['Country'].fillna(adult['Country']).dropna().mode()[0]
```

```python
np.sum(adult.isnull())
```

```python
from scipy.stats import stats
z= np.abs(stats.zscore(adult._get_numeric_data()))
print(z)
adult=adult[(z<3).all(axis=1)]
print(adult.shape)
```
